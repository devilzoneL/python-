```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set(style = 'darkgrid', rc = {'figure.figsize':(10, 6)}, font_scale = 1.5)
```

```python
sns.distplot(tips['total_bill'], hist = False, vertical= True, color = 'red')
plt.show()
```

```python
flights = sns.load_dataset('flights')
flights.head()
fp = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
fp.head()
sns.heatmap(fp)
plt.show()
```

```python
sns.heatmap(fp, cmap = 'magma', linecolor = 'white', linewidth = 1)
plt.show()
```

```python
sns.clustermap(fp, cmap = 'coolwarm')
plt.show()
```

```python
iris = sns.load_dataset('iris')
iris.head()
sns.PairGrid(iris)
plt.show()
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
plt.show()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
plt.show()
```

```python
g = sns.PairGrid(iris)
g.map_upper(plt.scatter)
plt.show()
```

```python
g = sns.PairGrid(iris)
g.map_lower(sns.kdeplot)
plt.show()
```

```python
g = sns.PairGrid(iris)

g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
plt.show()
```

```python
tips.head()
sns.set(font_scale = 1)
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker')
g.map(sns.histplot, 'total_bill')
plt.show()
```

```python
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker')
g.map(plt.scatter, 'total_bill', 'tip')
plt.show()
```
